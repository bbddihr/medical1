```python
import pandas as pd
df=pd.read_excel('../z20_data/score1.xlsx',index_col='지원번호')
df
```

```python
#합계라는 컬럼을 생성. 3학년 중에 성적이 가장 낮은 학생을 출력하시오.
```

```python
df['합계']=df['국어']+df['영어']+df['수학']+df['과학']+df['사회']
df
```

```python
df['학년']==3
df
```

```python
### 평균 컬럼을 추가해서 grade컬럼에 A>=90,B>=80,C>=70,D>=60,F<60학점을 출력하시오.

df['평균']=df['합계']/5
df
```

```python
def change(score):
    if score >=90: return 'A'
    elif score >=80: return 'B'
    elif score >= 70: return 'C'
    elif score >= 60 : return 'D'
    else: return 'F'
```

```python
df['grade']=df['평균'].apply(change)
```

```python
#SW특기 컬럼을 모두 대문자로 변경,NaN데이터는 모두 Python으로 입력하시오.
def upper_change(txt):
    
     if pd.isnull(txt):
         return "PYTHON"
     else:
         return txt.upper()
```

```python
df['SW특기']=df['SW특기'].apply(upper_change)
```
