```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

```python
df['국어']=df['국어'].astype(int)
```

```python
df
```

```python
import numpy as np
df.loc['3번','영어']=np.nan
df.loc['4번','수학']=np.nan
df.loc['5번',['국어','영어']]=np.nan
```

```python
df['총합']=0
df['총합']=df['국어']+df['영어']+df['수학']+df['과학']+df['사회']
df['총합'].sum()
```

```python
df.fillna(0,inplace=True)
```

```python
df['결과']='불합격'
df
```

```python
#총합 300 이상이면 합격,불합격
df.loc[df['총합']>300,'결과']='합격'
#400점 이상학생은 우수 
df.loc[df['총합']>400,'결과']='우수'
df
```

```python
#총합,결과 사회 뒤쪽에 옮겨서 출력
cols=list(df.columns)
df=df[cols[0:8]+cols[9:]+[cols[8]]]
df
```
