```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

```python
def add_score(kor):
    kor=kor+20
    if kor>100: kor=100
    return kor
```

```python

#함수사용해서
#국어점수를 20점더하고 100으로 변경
df['국어']=df['국어'].apply(add_score)
df
```

```python
#100점이상의 국어점수 100점으로 변경
df['국어']>100
df[df['국어']>100,'국어']=100
```

```python

df['결과']=df['평균'].apply(lambda score:'합격' if score>=65 else "재시험" if score==65 else '불합격')
```

```python
#(국어+영어+수학+과학+사회)-> 총합,평균의 칼럼 추가
#평점 컬럼 만들어서 90이상이면a.80점b,c,d, 90점 미만 F함수 사용해서 
```

```python
df['평점']=df['평균'].apply(grade)
```

```python
def grade(score):
    if score>=90 :return 'A'
    elif score>=80 :return 'B'
    elif score>=70 :return 'C'
    elif score>=60 :return 'D'
    else: return "F"
```
