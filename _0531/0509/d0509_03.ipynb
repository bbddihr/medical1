```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

```python
##  2개 이상 조건 검색부분

lang=['python','Java']
filt= df['SW특기'].isin(lang)
df[filt]
```

```python
#startwith,contains,lower,upper
#문자열 함수 적용-lower()
lang=['python','Java']
df['SW특기'].str.upper()
df['SW특기'].str.lower()
#python ==''
#2개 이상의 검색을 할수 있도록 함.isin명령어 사용
filt= df['SW특기'].str.lower().isin(lang)
df[filt]
```

```python
#java 라는 글자가 포함되어 있는 것을 검색
#na=False, Nan데이터를 True,False 처리해줌.
# filt= df["SW특기"].str.lower().contains('Java',na=True)
filt=df["SW특기"].str.lower().str.contains('java',na=False)
df[filt]
```

```python
import pandas as pd
df=pd.read_csv('../z20_data/stu_score1.csv',index_col='no')
df
```

### 이름에 i가 들어가 있는(대소문자 구분없이) 학생을 검색하시오.

```python
filt=df["name"].str.lower().str.contains('i')
len(df[filt])
df.loc[~filt,'id']
```
