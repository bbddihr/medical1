```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family']='Malgun Gothic'  #한글설정
matplotlib.rcParams['font.size']= 15   #글자크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
x=[1,2,3]
y=[2,4,8]

#graph그리기, 선두께 linewidth, label:범례
#plt.plot(x,y,linewidth=3,market='o',marketsize=10,markeredgecolor='blue', markerfacecolor='yellow',label='국어')
plt.plot(x,y,linewidth=3, marker='o',ms=10,mec='red', mfc='yellow',label='국어',alpha=0.7, ls=':')
# plt.plot(x,y,label='국어')
plt.legend(loc=(0.2,0.5))  #위치지점 변경
plt.title('막대그래프')
#y축 표시 간격을 조정
plt.yticks([5,7,9,11,13,15])
plt.xticks([1,2,3])
```
