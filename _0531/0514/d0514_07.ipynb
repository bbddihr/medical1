```python

import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family']='Malgun Gothic'  #한글설정
matplotlib.rcParams['font.size']= 15   #글자크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd
df=pd.read_excel('../z20_data/score1.xlsx',index_col='지원번호')
df.head(3)
```

```python
x=df.head(5)['이름']
y=df.head(5)['국어']
y2=df.head(5)['영어']
y3=df.head(5)['수학']
```

```python
plt.plot(x,y,label='국어',marker='D')
plt.plot(x,y2,label='영어',marker='o',linestyle='--')
plt.plot(x,y3,label='수학',marker='s',ls='-.')
#ncol 범례 1줄에 3개 까지
for i, txt in enumerate(y3):
    plt.text(x,y3[i],txt,ha='center')
plt.legend(loc='upper center',ncol=3)
plt.show()
```

```python
plt.bar(x,y3,label='국어')
colors=['r','g','b','y']
# plt.plot(x,y2,label='영어',marker='o',linestyle='--')
# plt.plot(x,y3,label='수학',marker='s',ls='-.')
#ncol 범례 1줄에 3개 까지
for i, txt in enumerate(y3):
    plt.text(x[i],y3[i]+0.9,txt,ha='center')
    plt.legend(loc='upper right]')
plt.legend(loc='upper center',ncol=3)
plt.show()
```

```python
plt.bar(x,y3,label='국어', width=0.2, alpha=0.5)

for i, txt in enumerate(y3):
    plt.text(x[i],y3[i]+0.9,txt,ha='center')
    
#패턴넣기
plt.legend(loc='upper right')
plt.show
```

```python
# 크기 조절
bar=plt.bar(x,y3,label='국어',width=0.2,alpha=0.5)
bar[0].set_hatch('/')
bar[1].set_hatch('x')
bar[2].set_hatch('..')
for i,txt in enumerate(y3):
    plt.text(x[i],y3[i]+0.9,txt,ha='center')
# 패턴넣기
plt.legend(loc='upper right')
plt.show()
```
