```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
matplotlib.rcParams['font.family']='Malgun Gothic'  #한글설정
matplotlib.rcParams['font.size']= 10  #글자크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd
df=pd.read_excel('../z20_data/score1.xlsx', index_col='지원번호')
df.head(3)
```

```python
plt.plot(df['이름'],df['영어'])
plt.plot(df['이름'],df['수학'])
```

```python
#산점도 그래프
plt.scatter(df['영어'],df['수학'])
plt.xlabel('영어점수')
plt.ylabel('수학점수')
```

```python
plt.scatter(df['영어'],df['수학'])
plt.scatter(df['영어'],df['과학'])
```

```python
np.random.rand(8)*1000
```

```python
sizes = np.random.rand(8)*1000
#s는 산점도 그래프 점의 크기 조절
plt.scatter(df['영어'],df['수학'],s=sizes)
```

```python
#학년별 크기조절 1학년*500, 2학년*500=1000, 3학년*500=1500
sizes=df['학년']*500
plt.scatter(df['영어'],df['수학'],s=sizes,c=df['학년'],cmap='viridis')
plt.xlabel('영어점수')
plt.ylabel('수학점수')
#컬러바 눈금표시 : ticks, 라벨표시: label, 컬러바 크기 조절: shrink, 
# 컬러바 위치: orientation 
plt.colorbar(ticks=[1,2,3],label='학년',shrink=0.5,orientation='horizontal')
```
