## dropna-Nan처리:지우기
axis : index,columns
how: any,all

```python
#Nan 데이터가 들어가있는 컬럼 전체 삭제
df.dropna(axis='columns', inplace=True)
df
```

```python
df["SW특기"].dropna(inplace=True)
```

```python
# Nan 들어가있는 컬럼에서 모든 컬럼이 Nan인것만 삭제
df.dropna(axis='columns', how='all', inplace=True)
```

```python
#Nan 들어가있는 컬럼에서 1개라도 Nan포함되어 있으면 삭제
df.dropna(axis='columns',how='any',inplace=True)
```

```python
#Nan 들어가있는 index에서 1개라도 Nan이 포함되어 있으면 삭제
df.dropna(axis='index',how='any',inplace=True)
```
