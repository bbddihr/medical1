```python
#t
#화면 사이즈 확대
plt.figure(figsize=(13,5))
plt.bar(df.index,df['출생아 수'])

#값 표시
for i ,txt in enumerate(df['출생아 수']):
    plt.text(df.index[i],df['출생아 수'][i]+0.5,txt,ha='center')
    
```

```python
plt.figure(figsize=(13,5))
plt.plot(df.index,df['합계 출산율'],marker='o')


for i ,txt in enumerate(df['합계 출산율']):
    plt.text(df.index[i],df['합계 출산율'][i]+0.05,txt,ha='center')
```

```python
#T
fig, ax1 = plt.subplots(figsize=(10,5))

# 1번째 막대 그래프
ax1.set_ylabel('출생아 수')
ax1.set_ylim(250,700)
ax1.set_yticks([300,400,500,600])
for i, txt in enumerate(df['출생아 수']):
    ax1.text(df.index[i],df['출생아 수'][i]+5,txt,ha='center')
ax1.bar(df.index,df['출생아 수'], color='#ff812d')

# 2번째 꺽은선 그래프
# 1개의 그래프에 내용을 2개 넣기
ax2 = ax1.twinx()
ax2.set_ylabel('합계 출산율')
ax2.set_ylim(0,1.5)
ax2.set_yticks([0,1])
ax2.plot(df.index, df['합계 출산율'],color='#ffd100',marker='o',markersize=10,markeredgecolor='w',markeredgewidth=3,linewidth=3)

# 데이터 출력
for i,txt in enumerate(df['합계 출산율']):
    ax2.text(df.index[i],df['합계 출산율'][i]+0.05,txt,ha='center')
```

```python
#두개를 하나그래프에 합침. subplots()
fig,ax1 = plt.subplots(figsize=(10,5))
fig.suptitle('출생아 수 및 합계출산율')

ax1.set_ylabel('출생아 수(천 명)')
ax1.set_ylim(250,700)
ax1.set_yticks([300,400,500,600])
for i, txt in enumerate(df['출생아 수']):
    ax1.text(df.index[i],df['출생아 수'][i],txt,ha='center')
ax1.bar(df.index,df['출생아 수'],color='#ff812d')
#twinx x축은 같이 사용할것임
ax2 = ax1.twinx()  
ax2.set_ylabel('합계 출산율(가임여성 1명당 명)')
ax2.set_ylim(0,1.5)
ax2.set_yticks([0,1])
for i, txt in enumerate(df['합계 출산율']):
    ax2.text(df.index[i].df['합계 출산율'][i]+1,txt,ha='center')
ax2.plot(df.index,df['합계 출산율'],color='#ffd100',marker='o',markersize=10,
         markeredgecolor='w',markeredgewidth=3,lw=3)

```
