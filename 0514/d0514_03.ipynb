```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family']='Malgun Gothic'  #한글설정
matplotlib.rcParams['font.size']= 15   #글자크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
x=[1,2,3]
y=[2,4,8]

#graph그리기, 선두께 linewidth, label:범례
#plt.plot(x,y,linewidth=3,market='o',marketsize=10,markeredgecolor='blue', markerfacecolor='yellow',label='국어')
plt.plot(x,y,linewidth=3, marker='o',ms=10,mec='red', mfc='yellow',label='국어',alpha=0.7, ls=':')
# plt.plot(x,y,label='국어')
plt.legend(loc=(0.2,0.5))  #위치지점 변경
plt.title('막대그래프')
```

```python

x=['han','kim','hong']
y=[20,40,80]
y2=[40,20,80]
y3=[80,90,95]
#범례 이름지정
plt.plot(x,y,label='국어')
plt.plot(x,y2,label='영어')
plt.plot(x,y3,label='수학')
#범례 표시, 위치설정, ncol
# plt.legend(loc='upper left')
plt.title('막대그래프')
# x,y축의 설명문구,위치,색상
plt.xlabel('국어',loc='center',color='red')  #left,center,right
plt.ylabel('이름',loc='top',color='#ff9100')  #top,center,bottom
# plt.legend(loc='upper right')
```
