```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family']='Malgun Gothic'  #한글설정
matplotlib.rcParams['font.size']= 15   #글자크기
matplotlib.rcParams['axes.unicode_minus'] = False

x=[1,2,3]
y=[2,4,8]
```

```python
plt.plot(x,y,label='국어',linestyle='--',marker='o')
plt.legend(loc='upper center')
plt.show()
```

```python
x=[1,2,3]
y=[2,4,8]

#graph그리기, 선두께 linewidth, label:범례
#plt.plot(x,y,linewidth=3,market='o',marketsize=10,markeredgecolor='blue', markerfacecolor='yellow',label='국어')
plt.plot(x,y,linewidth=3, marker='o',ms=10,mec='red', mfc='yellow',label='국어',alpha=0.7, ls=':')
# plt.plot(x,y,label='국어')
plt.legend(loc=(0.2,0.5))  #위치지점 변경
plt.title('막대그래프')
#y축 표시 간격을 조정
plt.yticks([5,7,9,11,13,15])
plt.xticks([1,2,3])
```

```python
import pandas as pd
df=pd.read_excel('../z20_data/score1.xlsx',index_col='지원번호')
df
```

```python
#키를 꺽은 선 그래프로 출력하시오.
#이름, 키
#y축표시 170,180,190,200,210
```

```python
x=df['이름']
y=df['키']

plt.plot(x,y,linewidth=3, marker='o',ms=10,mec='blue', mfc='yellow',label='키',alpha=0.7, ls=':')
plt.legend(loc='upper right')  #위치지점 변경
#y축 표시 간격을 조정
for i,txt in enumerate(y):
    plt.text(x[i],y[i]+1.3,txt,ha='center',color='blue')

plt.xticks(rotation=45)    
plt.yticks([170,180,190,200,210])
plt.ylim(165,210)
plt.title('학생키 그래프')
```

```python
x=df['이름']
y=df['키']

plt.barh(x,y,label='키')
plt.legend(loc='upper center')  #위치지점 변경
#y축 표시 간격을 조정
for i,txt in enumerate(y):
    plt.text(y[i]+1.3,x[i],txt,ha='center',color='blue')

# plt.xticks(rotation=45)    
# plt.yticks([170,180,190,200,210])
plt.xticks([165,170,180,190,200,210])
plt.xlim(165,210)
plt.title('학생키 그래프')
```
