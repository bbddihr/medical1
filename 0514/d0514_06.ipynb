#합계 컬럼을 생성해서 막대그래프, 옆으로 누운 그래프를 출력하시오.
#합계점수를 출력할것
#국어+영어+수학+과학+사회

```python
import pandas as pd
df=pd.read_excel('../z20_data/score1.xlsx',index_col='지원번호')
df
```

```python
df['합계']=df['국어']+df['영어']+df['수학']+df['과학']+df['사회']
df
```

```python

import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family']='Malgun Gothic'  #한글설정
matplotlib.rcParams['font.size']= 15   #글자크기
matplotlib.rcParams['axes.unicode_minus'] = False
x=df['이름']
y=df['합계']
```

```python
plt.bar(x,y,label='합계')
plt.legend(loc='upper right')  #위치지점 변경

for i,txt in enumerate(y):
    plt.text(x[i],y[i]+2.0,txt,ha='center',color='black')

plt.title('학생키 그래프')

plt.show()
```

```python
plt.barh(x,y,label='합계')
plt.legend(loc='upper right')  #위치지점 변경

for i,txt in enumerate(y):
    plt.text(x[i],y[i],txt,ha='center',color='black')

plt.title('학생                                                                                                                                                                                                                                                                                   키 그래프')
```
